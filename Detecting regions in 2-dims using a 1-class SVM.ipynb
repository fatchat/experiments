```python
import numpy as np
from sklearn import svm
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def make_dataset(typ):
    dataset_=[]
    if typ=='3':
        for i in range(100):
            dataset_.append([10+2*np.random.random(),0+np.random.random()])
            dataset_.append([0+2*np.random.random(),10+np.random.random()])
            dataset_.append([5+2*np.random.random(),7+np.random.random()])
    elif typ=='square':
        for x in range(100):
            for y in range(100):
                dataset_.append([x/100.,y/100.])
    return np.array(dataset_)
```

```python
dataset=make_dataset('square')
dataset_graph=dataset.transpose()
a=plt.scatter(dataset_graph[0], dataset_graph[1])
```

```python
def classify(trainingdata,nu,validate_on_training=False):
    classifier=svm.OneClassSVM(nu=nu)
    classifier.fit(trainingdata)
    if validate_on_training:
        validationset=trainingdata
    else:
        validationset=[]
        for x in range(10):
            for y in range(10):
                validationset.append([-0.1+x/10.0,-0.1+y/10.0])
    classes=classifier.predict(validationset)
    cls1=np.array([validationset[idx] for idx,cls in enumerate(classes) if cls==1])
    cls2=np.array([validationset[idx] for idx,cls in enumerate(classes) if cls==-1])
    if len(cls1) > 0 and len(cls2) > 0:
        a=plt.plot(cls1.transpose()[0], cls1.transpose()[1], 'bx', cls2.transpose()[0], cls2.transpose()[1], 'rx')
```

```python
classify(dataset,0.02)
```

```python
classify(dataset,0.2)
```
